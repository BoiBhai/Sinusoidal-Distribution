```python
import numpy as np
import scipy
from scipy.integrate import quad
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
### Hyperparameters
EPSABS=1e-5
EPSREL=1e-5
```

# Sinusoidal Distribution

```python
class sinu:
    def __init__(self):
        type = 'distribution'

    def sinecurve(self, z, s,k):
        return (np.sin(np.pi * z**s))**k

    def norm_const(self, s,k):
        integrand = lambda z: self.sinecurve(z, s,k)
        result, _ = quad(integrand, 0, 1, epsabs=EPSABS, epsrel=EPSREL)
        return result

    def pdf(self, x, s,k, a=0,d=1):
        z = (x-a)/d
        sinu_val = self.sinecurve(z, s,k)/self.norm_const(s,k)
        return sinu_val

    def cdf(self, x, s,k, a=0,d=1):
        integrand = lambda z: self.sinecurve(z, s,k)
        result, _ = quad(integrand, 0, (x-a)/d, epsabs=EPSABS, epsrel=EPSREL)
        return result/self.norm_const(s,k)
```

## Moments

```python
def base_rmom(n, s, k):
    integrand = lambda z: z**n * sinecurve(z, s, k)

    # Compute the integral to get the moment
    moment_value, _ = quad(integrand, 0, 1, epsabs=EPSABS, epsrel=EPSREL)

    norm_factor = norm_const(s, k)
    base_mom =  moment_value / norm_factor
    return base_mom

def base_cmom(n, s, k):
    mean = base_rmom(1,s,k)
    integrand = lambda z: (z-mean)**n * sinecurve(z, s, k)

    # Compute the integral to get the moment
    moment_value, _ = quad(integrand, 0, 1, epsabs=EPSABS, epsrel=EPSREL)

    norm_factor = norm_const(s, k)
    base_mom =  moment_value / norm_factor
    return base_mom
```

```python
def mean(s,k, a=0,d=1):
    return a + d * base_rmom(1, s,k)

def var(s,k, a=0, d=1):
    return d**2 * base_cmom(2, s,k)

def skew(s,k, a=0, d=1):
    return base_cmom(3, s,k)/(base_cmom(2, s,k))**(3/2)

def kurt(s,k, a=0, d=1):
    return base_cmom(4, s,k)/(base_cmom(2, s,k))**2 - 3
```
