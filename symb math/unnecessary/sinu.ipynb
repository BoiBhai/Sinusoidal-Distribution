```python
import math
from scipy import integrate
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
```

```python
class sinu:

	def curve(z, omega=1, chi=1, flip=False):
		if not flip:
			return math.sin(math.pi * z**omega) ** chi
		else:
			return math.sin(math.pi * (1-z)**omega) ** chi

	def pdf(x, alpha=0, delta=1, omega=1, chi=1, flip=False):
		z = (x-alpha)/delta
		def curve_inner(x, alpha=0, delta=1, omega=1, chi=1, flip=False):
			sinu.curve(omega, chi, flip)
		area = integrate.quad(sinu.curve(x), 0, 1)
		return (1/area * sinu.curve_inner(z))

	def cdf(x, alpha, delta, omega, chi):
		return integrate.quad(lambda x: sinu.pdf(x, alpha, delta, omega, chi), alpha, alpha+delta)
```

```python
class sinu:
    @staticmethod
    def curve(z, omega=1, chi=1, flip=False):
        if not flip:
            return math.sin(math.pi * z**omega) ** chi
        else:
            return math.sin(math.pi * (1-z)**omega) ** chi

    @staticmethod
    def pdf(x, alpha=0, delta=1, omega=1, chi=1, flip=False):
        z = (x - alpha) / delta
        def curve_inner(z, omega=omega, chi=chi, flip=flip):
            return sinu.curve(z, omega, chi, flip)
        area, _ = integrate.quad(curve_inner, 0, 1)
        return (1 / area * sinu.curve(z, omega, chi, flip))
```
